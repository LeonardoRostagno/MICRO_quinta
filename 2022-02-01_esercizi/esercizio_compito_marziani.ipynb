```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./marziani.csv")
df
```

```python
df.info()
```

```python
df_not_NaN = df.dropna()

if df.size != df_not_NaN.size :
    print("Valori NaN trovati")
    df.dropna(inplace = True)
else:
    print("Valori NaN non trovati")
```

```python
specie = df["specie"].values
colore = df["colore"].values
n_arti = df["n_arti"].values
peso = df["peso"].values
altezza = df["altezza"].values
larghezza = df["larghezza"].values

arr_marz = [specie, colore, n_arti, peso, altezza, larghezza]
```

```python
fig, axs = plt.subplots(nrows = 5, ncols = 5, figsize = (25, 25))

for a in range(0, len(arr_marz)-1):
    for b in range(0, len(arr_marz)-1):
        if a == b:
            axs[a, b].hist(arr_marz[b][(specie == "Robby")])
            axs[a, b].hist(arr_marz[b][(specie == "Simmy")], alpha = 0.9)
            axs[a, b].set_title(f"{df.columns[a]} - {df.columns[b]}")
        else:
            axs[a, b].scatter(arr_marz[b][(specie == "Robby")], arr_marz[a][(specie == "Robby")])
            axs[a, b].scatter(arr_marz[b][(specie == "Simmy")], arr_marz[a][(specie == "Simmy")])
            axs[a, b].set_title(f"{df.columns[a]} - {df.columns[b]}")

plt.show()
```
