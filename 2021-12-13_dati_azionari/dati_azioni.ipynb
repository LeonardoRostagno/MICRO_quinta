```python
import numpy as np
import matplotlib.pyplot as plt
```

### Lettura File

```python
diz = {} # dizionario 
```

```python
def reader(nome_file):
    # creo liste ausiliarie
    list_data = [] # data
    list_AAPL = [] # Apple
    list_MSFT = [] # Microsoft
    list_GOOG = [] # Google
    list_AMZN = [] # Amazon
    list_FB = [] # Facebook
    list_TSLA = [] # Tesla
    
    # apro il file e prendo il testo
    file = open(nome_file, "r") # file neve - diga del chiodas
    testo = file.readlines()[1:][:-1]
    file.close()
    
    # inserisco gli elementi nelle liste ausiliarie
    for riga in testo:
        lista = riga.replace(" ", "").replace("\n", "").split(",")

        list_data.append(lista[0])
        list_AAPL.append(lista[1])
        list_MSFT.append(lista[2])
        list_GOOG.append(lista[3])
        list_AMZN.append(lista[4])
        list_FB.append(lista[5])
        list_TSLA.append(lista[6])
        
    # creazione array numpy
    arr_data = np.array(list_data,)
    arr_AAPL = np.array(list_AAPL, dtype = float)
    arr_MSFT = np.array(list_MSFT, dtype = float)
    arr_GOOG = np.array(list_GOOG, dtype = float)
    arr_AMZN = np.array(list_AMZN, dtype = float)
    arr_FB = np.array(list_FB, dtype = float)
    arr_TSLA = np.array(list_TSLA, dtype = float)
    
    # inserisco nei dizionari gli array
    diz["data"] = arr_data
    diz["AAPL"] = arr_AAPL
    diz["MSFT"] = arr_MSFT
    diz["GOOG"] = arr_GOOG
    diz["AMZN"] = arr_AMZN
    diz["FB"] = arr_FB
    diz["TSLA"] = arr_TSLA
```

```python
reader("dati_azioni.csv")
```

### Punto 1
calcolare il numero di giorni in cui Alphabet va meglio di Tesla

```python
np.count_nonzero(diz["GOOG"] > diz["TSLA"])
```

### Punto 2
calcolare i massimi locali nella percentuale giornaliera di Meta Platforms

```python
arr_prec = np.roll(diz["FB"], -1)[1: -1]
arr_succ = np.roll(diz["FB"], 1)[1: -1]

arr_FB = diz["FB"][1: -1]
arr_FB[np.logical_and(arr_FB > arr_prec, arr_FB < arr_succ)]
```

### Punto 3
calcolare se per Microsoft Corporation i giorni positivi sono più di quelli negativi

```python
pos = np.count_nonzero(diz["MSFT"] >= 0)
neg = np.count_nonzero(diz["MSFT"] < 0)
```

```python
if pos >= neg:
    print("Giorni positivi maggiori")
else:
    print("Giorni negativi maggiori")
```

### Punto 4
calcolare quanti sono i giorni in cui Microsoft Corporation è positiva e Apple è negativa

```python
np.count_nonzero(np.logical_and(diz["MSFT"] > 0, diz["AAPL"] < 0))
```

### Punto 5
stampare in quali date la somma delle variazioni di prezzo di Tesla e Apple è maggiore di 0

```python
diz["data"][(diz["TSLA"] + diz["AAPL"]) > 0]
```

### Punto 6
trovare se ci sono periodi di 3 giorni consecutivi in cui Amazon.com è negativa e in caso affermativo in che date iniziano

```python
var_prec1 = np.roll(diz["AMZN"], -1)[: -2]
var_prec2 = np.roll(diz["AMZN"], -2)[: -2]
var_attuale = diz["data"][: -2]

var_attuale[np.logical_and(np.logical_and(var_prec1 < 0, var_prec2 < 0), diz["AMZN"][: -2] < 0)]
```

### Punto 7
fare un grafico che rappresenti l'andamento delle  variazioni percentuali giornaliere  di Amazon.com

```python
fig, ax = plt.subplots(figsize = (50, 25))
ax.plot(diz["data"], diz["AMZN"], "b")
plt.show()
```

### Punto 8
fare un istogramma che rappresenti le  variazioni percentuali giornaliere  di Tesla.

```python
fig, ax = plt.subplots(figsize = (50, 25))
ax.hist(diz["TSLA"], bins = diz["TSLA"].size)
plt.show()
```
