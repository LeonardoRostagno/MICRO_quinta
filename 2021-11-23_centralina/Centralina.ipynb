```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
media = 19
varianza = 5
n_dati = 1000
val_err = 9999.9
n_dati_err = 20
n_colonne = 50
```

```python
dati = np.array(np.random.normal(media, np.sqrt(varianza), n_dati))
rand = np.random.choice(range(n_dati), n_dati_err, replace = False) 

dati[rand] = val_err
print(dati)
```

```python
val_err = np.std(dati)
print(val_err)
```

```python
pdf, bins = np.histogram(dati, n_colonne, density = True)
row_succ = len(bins)
# Calcola i centri dei bins
x_col = 0.5 * (bins[0 :row_succ - 1] + bins[1 : row_succ])
print("Deviazione standard: ", moment(2, x_col, pdf))
print("Valore teorico: ", varianza)
print("Skewness: ", moment(3, x_col, pdf))
print("Valore teorico: ", 0.0)
print("Flatness: ", moment(4, x_col, pdf))
print("Valore teorico: ", 3.0 * varianza**2)
```

```python
import scipy.optimize

def F(x, media, varianza):
    ff = -0.5 * np.log(2.0 * np.pi) - np.log(varianza) - ((x - media)**2.0) / (2.0 * varianza * varianza)
    return ff

# Limita l'intervallo del fit ai valori per cui y_i != 0 per evitare problemi con il log!
imin = 2
imax = -5
xdata = x_col[imin : imax]
ydata = np.log(pdf[imin : imax])
par, corr = scipy.optimize.curve_fit(F, xdata, ydata)
print("Parametri del fit [media, varianza]: ", par)
```
