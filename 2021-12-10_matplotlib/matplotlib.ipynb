```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.random.uniform(0, 100, 100)
a
```

```python
# grafico in modalit OO 
fig, ax = plt.subplots(figsize = (8, 4))

ax.plot(a) # asse x mette indice, in y i valori
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.cos(x)

fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = 1/x

fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(x, y)
plt.show()
```

```python
date = np.array(["01/01/2021", "02/01/2021", "03/01/2021", "04/01/2021"])
val = np.array([3.4, 5.1, 1.3, 2.8])

fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(date, val, "r--")
plt.show()
```

```python
n_lanci = 10000
dado1 = np.random.randint(1, 7, n_lanci)
dado2 = np.random.randint(1, 7, n_lanci)

somma = dado1 + dado2
```

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax.hist(dado1, bins = 6)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax.hist(dado2, bins = 6)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax.hist(somma, bins = 11)
plt.show()
```
