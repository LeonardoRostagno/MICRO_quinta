```python
dizionario_freq = {}
dizionario_seq = {}
```

```python
file = open("./sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt", "r")

# conta_frequenze
for row in file.readlines():
    lista = row[:-1].split(" ")
    
    for parola in lista:
        parola = parola.lower()
        
        if parola not in dizionario_freq.keys():
            if parola != "":
                dizionario_freq[parola] = 1 
        else:
            dizionario_freq[parola] += 1

print(dizionario_freq)

file.close()
```

```python
num = int(input("Inserire il numero di parole da voler visualizzare: "))

max_word = {}

# elenco parole + frequenti
for _ in range(num):
    max_freq = 0
    max_parola = ""
    
    for parola, valore in dizionario_freq.items():
        if valore > max_freq:
            max_freq = valore
            max_parola = parola
            
    del dizionario_freq[max_parola]
    max_word[max_parola] = max_freq      

print(f"le parole più frequenti sono:{max_word}")    
```

```python
file = open("./sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt", "r")

# conta_sequenze
for row in file.readlines():
    lista_row = row[:-1].split(" ")
    
    for parola in lista_row:
        parola = parola.lower()
        
        for i, word in enumerate(lista_row):
            if i < len(lista_row) -1:
                if lista_row[i+1] not in dizionario_seq :
                    dizionario_seq[word] = lista_row[i+1]
                    
print(dizionario_seq)

file.close()
```
