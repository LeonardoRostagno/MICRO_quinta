```python
n = 30
x = np.linspace(0, 2*np.pi, n)
y = np.sin(x)
z = np.random.normal(0, 1, n)
```

```python
# fig Ã¨ genitore di tutta la storia, ax sono gli assi destinati al grafico
fig, ax = plt.subplots(figsize = (10, 5)) # creato gli oggetti fig e ax, con una dimensione in pollici 

ax.plot(x, y) # .plot unisce i punti, consente anche le + svariate modifiche
ax.plot(x, z, 'r')

# plt.savefig("./nomefile.png", dpi = 200) --> consente di salvare il grafico ("nome file", risoluzione). 
# Salva come .jpg, .png, .eps(vettoriale non compresso), .pdf(vettoriale compresso)
plt.show() # mostra il grafico
```

Visualizzare 5 rette con i coefficienti angolari contenuti in lista, e tutte 5 nello stesso grafico, seguendo la formula y = m*x

```python
lista = [0, 1/2, 1, 2, 4]

x = np.linspace(0, 1, len(lista))
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
    
for i in range(len(lista)):
    ax.plot(x, lista[i]*x)

plt.show()
```
