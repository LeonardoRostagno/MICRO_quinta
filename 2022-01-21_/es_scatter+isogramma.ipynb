```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.uniform(0, 1, 100)
y = np.random.uniform(0, 1, 100)
size = np.random.uniform(150, 5, 100)
col = np.random.choice(['r', 'g', 'b', 'y'], 100)
```

Usando .scatter possiamo modificare maggiormente le proprietà dei punti, rendendoo più efficacie rispetto a .plot

```python
fig, ax = plt.subplots(figsize = (5, 5))
ax.scatter(x, y, s = size, c = col, alpha = 0.3) # alpha modifica l'evanescenza
plt.show()
```

Generiamo un istogramma, che seve a vedere le occorrenze degli eventi

```python
fig, ax = plt.subplots(figsize = (5, 5))
ax.hist(col)
plt.show()
```
