```python
import matplotlib.pyplot as plt
import numpy as np
```

## Matrici
- .imshow() consente di visualizzare matrici (ovvero anche le immagini) dove ogni valore Ã¨ indicativamente un pixel.
- cmap consente di modificare le scale di colore
- vmin e vmax comsentono di modificare gli estremi analoghi dei valoei dei pixel

```python
matrice = np.random.uniform(0, 1, (100, 100))
```

```python
fig, ax = plt.subplots(figsize = (5, 5))
ax.imshow(matrice, cmap = "gist_ncar")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
ax.imshow(img)
plt.show()
```

## + grafici nella stessa figura
- axs Ã¨ un array di assi
- devo anche modificare il numero di righe e colonne con nrows e ncols

```python
x = np.random.uniform(0, 1, 100)
y = np.random.uniform(0, 1, 100)
size = np.random.uniform(150, 5, 100)
col = np.random.choice(['r', 'g', 'b', 'y'], 100)
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10))

axs[0, 0].scatter(x, y, s = size, c = col, alpha = 0.3)
axs[0, 1].imshow(img)
axs[1, 0].imshow(matrice, cmap = "gist_ncar")
axs[1, 1].hist(col)

plt.show()
```
