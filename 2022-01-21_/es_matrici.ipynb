```python
import matplotlib.pyplot as plt
import numpy as np
```

Usare numpy per creare una matrice 100 x 100 di numeri interi compresi tra 0 e 255. Crea una figura con 2 grafici:
- il primo rappresenta la matrice con la mappa di colore che preferisci
- il secondo rappresenta la distribuzione dei valori degli elementi della matrice usando un istogramma

```python
matrice = np.random.uniform(0, 255, (100, 100))
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 10))

axs[0].imshow(matrice, cmap = "gist_ncar")
axs[1].hist(matrice.ravel(), bins = 256)

plt.show()
```
